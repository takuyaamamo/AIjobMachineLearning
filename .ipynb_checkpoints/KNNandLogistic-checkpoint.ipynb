### データ準備

```python
# 乳がんデータを読み込む
import pandas as pd
from IPython.core.display import display
from sklearn.datasets import load_breast_cancer

dataset = load_breast_cancer()
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = pd.Series(dataset.target, name='y')

# データを確認
print('------------------------------------------------')
print(f'X shape: {X.shape}')
print('------------------------------------------------')
print(y.value_counts())
print('------------------------------------------------')
print('y = 0 meas Marignat(悪性), y = 1 means Benign(良性):')
display(X.join(y).head())
```

### 評価指標の呼び出し

```python
# 正解率（Accuracy）を使用する
from sklearn.metrics import accuracy_score
```

### 学習

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)

pipelines = {
    'knn': Pipeline([('scl', StandardScaler()),
                     ('est', KNeighborsClassifier())]),
    'Logistic': Pipeline([('scl', StandardScaler()),
                          ('est', LogisticRegression(random_state=1))])
}

scores = {}
for pipe_name, pipeline in pipelines.items():# items(): 各要素のキーkeyと値valueに対してforループ処理
    pipeline.fit(X_train, y_train)
    print(pipe_name, ':Fitting Done')
    # 性能（正解率を比較する）
    scores[(pipe_name, 'train')] = accuracy_score(y_train, pipeline.predict(X_train))
    scores[(pipe_name, 'test')] = accuracy_score(y_test, pipeline.predict(X_test))
    
pd.Series(scores).unstack()
```
