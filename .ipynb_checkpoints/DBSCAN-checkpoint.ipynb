# DBSCANの実装
- 密度ベースの教師なし学習、Core点（半径ε以内に指定個数minPts以上の隣接点があるような点）、Border点（半径ε以内の隣接点がminPts未満だが、コア点の半径εに位置する点）、Noise点（それ以外）に分け、半径ε以内のCore点同士を結ぶことでクラスタ化する。

### データの準備

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, labels_true = make_moons(n_samples=1000,#サンプル数
                           noise=0.05,
                           random_state=0)

print('Generated data X:')
print(X[:5])
print('Generated data y:')
print(labels_true[:5])

# hstackで横に結合、行,列=(labels_trueの個数行,1列)にreshapeしている
df = pd.DataFrame(np.hstack([X, labels_true.reshape(len(labels_true), 1)]),
                 columns=['X0', 'X1', 'label'])

col = df.label.map({0:'g', 1:'b'})
df.plot(x='X0', y='X1', kind='scatter', c=col, colorbar=False, figsize=(6,4))
```

### DBSCANの実装

```python
from sklearn.cluster import DBSCAN
# DBSCANのインスタンス作成,eps:探索範囲, min_samples:epsの中にいくつ点が存在していたらクラスタだとみなすか,metric:インスタンス間距離の計算方法
db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
fig2, ax2 = plt.subplots()

# 学習と同時に予測も行う
for j in range(3):
    ax2.plot(X[db.fit_predict(X) == j, 0],
            X
            [db.fit_predict(X) == j, 1], 'o',
            label='series ' + str(j))
```
