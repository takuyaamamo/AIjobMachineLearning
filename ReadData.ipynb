# サンプルデータの読み込み

### ロードボストンデータの読み込み

```python
# Boston house-price datase for Regression（回帰）
import pandas as pd #データ加工ライブラリpdとして読み込む
from sklearn.datasets import load_boston
from IPython.core.display import display #表の表示機能の読み込み
```

```python
# set dataframe
dataset = load_boston()
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
display(X.tail())
# targetをMEDVとして取り出す
y = pd.Series(dataset.target, name='MEDV')# Median value of owner-occupied homes in $1000's
display(y.tail())
```

```python
print('--------------------------------------')
print('X shape:(%i, %i)'%X.shape)# shapeで行列数を取り出せる
print('--------------------------------------')
print(y.describe())# describeで統計値の取り出し
display(X.join(y).head())# きれいに強制出力してくれる
```

### CSVからの読み出し

```python
# big mart sales amount for Regression
# https://datahack.analyticsvidhya.com/contest/practice-problem-big-mart-sales-iii/
import pandas as pd
import numpy as np
df = pd.read_csv('./data/av_big_mart_sales_UWu5bXk.csv',
                 header=0)  # headerなしの場合はNone
X = df.iloc[:,:-1]  # 全行対象,最終カラム以外をXとする
y = df.iloc[:,-1] # 全行対象,最終カラムをyとする

# check the shape
print('----------------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('----------------------------------------')
print(y.describe())
# 目的変数に欠損が含まれているとアルゴリズムが学習できないためチェック
print('----------------------------------------')
print('Check null count of target variable: %i' % y.isnull().sum())
print('----------------------------------------')
display(X.join(y).head(5))
```
