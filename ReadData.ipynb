# サンプルデータの読み込み

### ロードボストンデータの読み込み

```python
# Boston house-price datase for Regression（回帰）
import pandas as pd #データ加工ライブラリpdとして読み込む
from sklearn.datasets import load_boston
from IPython.core.display import display #表の表示機能の読み込み
```

```python
# set dataframe
dataset = load_boston()
# 複数次元のデータの場合DataFrame()を使用する
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
display(X.tail())
# targetをMEDVとして取り出す、1個の場合Series()を使用する
y = pd.Series(dataset.target, name='MEDV')# Median value of owner-occupied homes in $1000's
display(y.tail())
```

```python
print('--------------------------------------')
print('X shape:{}'.format(X.shape))# shapeで行列数を取り出せる
print('--------------------------------------')
print(y.describe())# describeで統計値の取り出し
display(X.join(y).head())# きれいに強制出力してくれる
```

### CSVからの読み出し

```python
# big mart sales amount for Regression
# https://datahack.analyticsvidhya.com/contest/practice-problem-big-mart-sales-iii/
import pandas as pd
import numpy as np

df = pd.read_csv('./data/av_big_mart_sales_UWu5bXk.csv',
                 header=0)  # headerなしの場合はNone
X = df.iloc[:, :-1]  # 全行対象,最終カラム以外をXとする
y = df.iloc[:, -1] # 全行対象,最終カラムをyとする

# check the shape
print('----------------------------------------')
print('X shape: (%i,%i)' %X.shape)
print('----------------------------------------')
print(y.describe())
# 目的変数に欠損が含まれているとアルゴリズムが学習できないためチェック
print('----------------------------------------')
print('Check null count of target variable: %i' %y.isnull().sum())
print('----------------------------------------')
display(X.join(y).head())
```

### Irisの読み込み

```python
# Iris dataset for Classification
# Classic and very easy multi-class classification dataset
import pandas as pd
from sklearn.datasets import load_iris

dataset = load_iris()

# Set dataframe
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = pd.Series(dataset.target, name='y')
# check the shape
print('----------------------------------')
print(f'X shape:{X.shape}')
print('----------------------------------')
#value_counts()はカテゴリーの数でまとめてくれる
print(y.value_counts())
display(X.join(y).head())
print(dataset.DESCR)
```

### Breast cancer wisconsin datasetの読み出し

```python
# Breast cancer wisconsin dataset for Classification
import pandas as pd
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()

#set dataframe
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
y = pd.Series(dataset.target, name='y')

#check the shape
print('------------------------------------------')
print(f'X shape:{X.shape}')
print('------------------------------------------')
print(y.value_counts())
print('------------------------------------------')
display(X.join(y).head())
```

### Loan screening CSVの読み込み

```python
# Loan screening data for Classification 
import pandas as pd

df = pd.read_csv('./data/av_loan_u6lujuX_CVtuZ9i.csv', header=0)
print(df.iloc[:0,:])
print(df.shape)
X = df.iloc[:,:-1]
print(X.shape)
y = df.iloc[:,-1]
print(y.shape)
```

```python
# check the shape
print('----------------------------------------')
print(f'X shape:{X.shape}')
print('----------------------------------------')
print(y.value_counts())
print('----------------------------------------')
print(f'Check null count:{y.isnull().sum()}')
print('----------------------------------------')
```

```python
# Converting string to number
# 機械学習において文字列は理解できないため数字に変換する
class_mapping = {'N':1, 'Y':0}
print(y.tail())
# Nを1に,Yを0に変えている
y = y.map(class_mapping)
print(y.tail())
# check the converted result
print('check the result converted')
print(y.value_counts())
display(X.join(y).head())# headのdefoultは5
```
